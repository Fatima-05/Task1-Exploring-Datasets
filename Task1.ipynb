```python
#importing libraries
import os
import pandas as pd
import seaborn as sns
import kagglehub
import matplotlib.pyplot as plt
```

```python
#loading the dataset using url path
path = "https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
df = pd.read_csv(path)
```

```python
#using head to show first few entries of the dataset
df.head()
```

```python
#using shape to find the shape (rows and columns) of the dataset
shape=df.shape
print("Shape of iris dataset: ",shape)
```

```python
#printing the names of the columns in the dataset in the form of a list
column_names=df.columns.tolist()
print("Column names in iris dataset: ",column_names)
```

```python
#printing structural info of the dataset using info
print(df.info())
```

```python
#printing statistical info of the dataset using describe
print(df.describe())
```

```python
#using scatter plot to show the relationship between the features
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal.length', y='sepal.width', hue='variety')
plt.title("Sepal Length vs Width")
plt.show()
```
